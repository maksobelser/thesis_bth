```python
import matplotlib.pyplot as plt

# Assuming 'normal' and 'exponentiated' are your np arrays.

# Filter out the zeros
normal_non_zero = normal[normal != 0]
exponentiated_non_zero = exponentiated[exponentiated != 0]

# Plotting only the non-zero values
plt.figure(figsize=(10, 6))
plt.hist(exponentiated_non_zero, bins=50, alpha=0.5, label='Exponentiated', log=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Non-Zero Values for Exponentiated matrix')
plt.savefig("fig_251023/hist_of_non_0_values.png")
plt.show()
```
